```python
import yaml
from typing import List
```

```python
def get_yaml(source: str):

    with open(source, "r") as stream:
        try:
            y = yaml.safe_load(stream)
        except yaml.YAMLError as exc:
            print(exc)
    return y
```

```python
to_convert = [
    get_yaml(i) for i in [
    'Agents_cards_v1.yaml', 
    'Group_cards_v1.yaml', 
    'Objective_cards_v1.yaml', 
        ]
    ]
```

```python
def replace_wrong_symbols(string: str) -> str:
    string = string.lower().replace(' ', '_').replace('(', '').replace(')', '')
    return string 
```

```python
converted = []
for i in to_convert:
    conv_in = []
    for j in i:
        d_in = {}
        for key in j.keys():
            d_in[replace_wrong_symbols(key)] = j[key]
        conv_in.append(d_in)
    converted.append(conv_in)
```

```python
converted = {
    'agents_cards': converted[0],
    'group_cards': converted[1],
    'objective_cards': converted[2],
}
```

```python
def split_bias(string: str) -> List[str]:
    return string.split(", ")
```

```python
for i in converted['objective_cards']:
    i['bias_icons'] = split_bias(i['bias_icons'])
    if i['special_ability'] == 'None':
        i['special_ability'] = None
```

```python
with open('converted.yaml', "w") as stream:
    yaml.dump(converted, stream)
```
